```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("remoteok_jobs.xlsx")
df.head()
```

```python
df.shape
df.info()
df.isna().sum()
```

```python
#Removeing Duplicates
df = df.drop_duplicates(subset="Job URL")
df.head()
```

```python
#fixing missing values
df["Location"] = df["Location"].fillna("Remote / Not Specified")
df["Job Type"] = df["Job Type"].fillna("Not Specified")
df["Skills / Tags"] = df["Skills / Tags"].fillna("Unknown")
df.head()
```

```python
#cleanfing text
df["Job Title"] = df["Job Title"].str.replace("&amp;", "&", regex=False)
df["Skills / Tags"] = df["Skills / Tags"].str.lower()
df.head()
```

```python
df["Date Posted"] = pd.to_datetime(df["Date Posted"])
df.head()
```

```python
#train test split
df["Skill_List"] = df["Skills / Tags"].str.split(",")
print(df["Skill_List"])
```

```python
#Analyzing top job roles
df["Job Title"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Job Titles")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
#jobs by location
df["Location"].value_counts().head(10).plot(kind="bar")
plt.title("Top Job Locations")
plt.show()
```
